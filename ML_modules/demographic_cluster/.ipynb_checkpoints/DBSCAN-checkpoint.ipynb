```python
import pandas as pd
import requests
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
# Fetch data from the API endpoint
url = "http://127.0.0.1:5000/api/v1.0/demographic_data_2022_budget"
response = requests.get(url)
data = response.json()
```

```python
# Convert data to DataFrame
demographic_df = pd.DataFrame(data)
demographic_df.head()
```

```python
# Select features for clustering
features = ["2022 Budget", "Population density per square kilometre"]
X = demographic_df[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform Agglomerative Hierarchical Clustering
# You can adjust parameters like linkage and number of clusters
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg_clustering.fit(X_scaled)

# Add cluster labels to the DataFrame
demographic_df['Cluster'] = agg_clustering.labels_
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
for cluster in demographic_df['Cluster'].unique():
    plt.scatter(demographic_df[demographic_df['Cluster'] == cluster]['Population density per square kilometre'], 
                demographic_df[demographic_df['Cluster'] == cluster]['2022 Budget'],
                label=f'Cluster {cluster}', alpha=0.7)
    for i, ward in enumerate(demographic_df[demographic_df['Cluster'] == cluster]['Ward']):
        plt.text(demographic_df[demographic_df['Cluster'] == cluster]['Population density per square kilometre'].iloc[i], 
                 demographic_df[demographic_df['Cluster'] == cluster]['2022 Budget'].iloc[i], 
                 ward, fontsize=8)

plt.title('Agglomerative Hierarchical Clustering of Wards')
plt.xlabel('Population density per square kilometre')
plt.ylabel('2022 Budget')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Select features for clustering
features = ["Population density per square kilometre", "2022 Budget"]
X = demographic_df[features]

# Perform hierarchical clustering
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(Z, labels=demographic_df["Ward"].values, leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Wards')
plt.ylabel('Distance')
plt.show()
```
